## **1. Atkore Inc. (ATKR)**

```python
class BootCampTask(QCAlgorithm):

    def Initialize(self):
        self.SetStartDate(2019, 5, 3)
        self.SetEndDate(2021, 5, 3)

        #1,2. Select IWM minute resolution data and set it to Raw normalization mode
        self.atkr = self.AddEquity("ATKR", Resolution.Minute)
        self.atkr.SetDataNormalizationMode(DataNormalizationMode.Raw)

    def OnData(self, data):

        #3. Place an order for 100 shares of IWM and print the average fill price
        if not self.Portfolio.Invested:
            self.MarketOrder("ATKR", 100)
            self.Debug( str(self.Portfolio["ATKR"].AveragePrice))
```
